# Anomaly Detection

Outliers for a cluster

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.neighbors import NearestNeighbors
from sklearn.datasets import load_iris
import seaborn as sns
from pyod.models.iforest import IForest
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from pyod.models.abod import ABOD
```

```python
sns.set_style('darkgrid')
```

```python
iris = load_iris()

# X = iris.data
# y = iris.target
```

```python
plt.scatter(x=iris.data[:,:1], y=iris.data[:,1:2])
plt.xlabel('Sepal Width')
plt.ylabel('Sepal Length')
```

```python
# Build the KNN model
KNN = NearestNeighbors(n_neighbors=3)
KNN.fit(iris.data)
```

```python
# determine the cutoff point

# calculate average distances
distances, indices = KNN.kneighbors(iris.data)
plt.plot(distances.mean(axis=1))
```

```python
distances, indices = KNN.kneighbors(iris.data)
distances
```

```python
# plot the cutoff point

plt.plot(distances.mean(axis=1))
plt.axhline(y=0.28, color="r")
```

```python
anomaly_indices = np.where(distances.mean(axis=1) > 0.28)
anomaly_indices
```

```python
df = pd.DataFrame(iris.data, columns=iris.feature_names)
```

```python
anomaly_values = df.iloc[anomaly_indices]
anomaly_values
```

```python
plt.scatter(df['sepal width (cm)'], df['sepal length (cm)'])
plt.scatter(anomaly_values['sepal width (cm)'], anomaly_values['sepal length (cm)'])
```

## Using PyOD

```python
# Data prep for iForest

# Standardize
S_Scaler = StandardScaler()
X_scaled = S_Scaler.fit_transform(iris.data)
```

```python
# Apply PCA (for visualization only)
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)
X_pca
```

### Isolation Forest

It's a tree-based ensemble algorithm. It isolates outliers by creating partitions (branches).

Recommended to scale data before using isolation forest

```python
# Create the model
IF = IForest(random_state=10)

# Fit
IF.fit(X_scaled)
```

```python
# Apply model to predict anomalies
y_pred = IF.predict(X_scaled)
```

```python
# Plotting
plt.scatter(X_pca[:,0], X_pca[:,1], c=y_pred, cmap='bwr')
```

## ABOD Algorithm

Angle-Based Outlier Detection

```python
model = ABOD()
```

```python
# Fit the model
model.fit(X_scaled)

# Apply prediction
y_pred_abod = model.predict(X_scaled)
```

```python
plt.scatter(X_pca[:,0], X_pca[:,1], c=y_pred_abod, cmap='bwr')
```
